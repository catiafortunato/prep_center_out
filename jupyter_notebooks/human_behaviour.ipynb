# Straigh Lines

```python
dir = '/home/cf620/Documents/curved_manifold/data/human/'
```

```python
lines_files = 't5_2019_06_03_straightLines_trialdata.mat'
```

```python
df = mat2dataframe(dir + lines_files, True)
```

```python
df['bin_size'] = 0.01
df['movement_on'] = 50

df = restrict_to_interval(df, start_point_name='movement_on', rel_start=-2, rel_end=100)
```

```python
df['trial_id'] = df.index
```

```python
#df = transform_signal(df, "M1_spikes", 'sqrt')
```

```python
df = transform_signal(df, "M1_spikes", 'zero_normalize')
df = add_firing_rates(df, 'smooth')
```

```python
td_avg = trial_average(df, "direction")
```

```python
td_avg
```

```python
fig, axs= plt.subplots(1, 4, figsize=(15, 6))

k=0
for i in range(1,17,4):
    axs[k].imshow(td_avg.loc[str(i),'M1_rates'].T)
    k+=1

plt.suptitle('Straight Lines', fontsize=16)

plt.savefig('/home/cf620/Documents/curved_manifold/plots/human_lines_rates.pdf')
```

# Letters

```python
letters_files = 't5_2020_01_15_singleLetters_trialdata.mat'
```

```python
df = mat2dataframe(dir + letters_files, True)
```

```python
df['bin_size'] = 0.01
df['movement_on'] = 50

df = restrict_to_interval(df, start_point_name='movement_on', rel_start=-2, rel_end=100)
```

```python
df['trial_id'] = df.index
```

```python
df = transform_signal(df, "M1_spikes", 'zero_normalize')
df = add_firing_rates(df, 'smooth')
```

```python
td_avg = trial_average(df, "letter")
```

```python
td_avg.index
```

```python
fig, axs= plt.subplots(1, 4, figsize=(15, 6))

list_letters = ['a', 'f', 'm', 'x']
k=0
for i in list_letters:
    axs[k].imshow(td_avg.loc[i,'M1_rates'].T)
    axs[k].set_title(i)
    k+=1

plt.suptitle('Letters', fontsize=16)

plt.savefig('/home/cf620/Documents/curved_manifold/plots/human_letters_rates.pdf')
```
