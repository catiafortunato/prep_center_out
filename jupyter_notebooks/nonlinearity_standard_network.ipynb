```python
import matplotlib.pyplot as plt
import numpy as np
import glob, os
import scipy.io as sio
from scipy.signal import savgol_filter
import random
from scipy import stats
```

```python
pca_m10 = np.zeros((8,10))
pca_m40 = np.zeros((8,10))
pca_m70 = np.zeros((8,10))
pca_m100 = np.zeros((8,10))

isomap_m10 = np.zeros((8,10))
isomap_m40 = np.zeros((8,10))
isomap_m70 = np.zeros((8,10))
isomap_m100 = np.zeros((8,10))
```

```python
data_dir = '/home/cf620/git/prep_center_out/simulation/results/'

for i in range(1,9):

    pca_m10[i-1,:] = np.array(sio.loadmat(data_dir + '10002'+str(i)+'/1/pca_mean.mat')['dist'])
    pca_m40[i-1,:] = np.array(sio.loadmat(data_dir + '10002'+str(i)+'/2/pca_mean.mat')['dist'])
    pca_m70[i-1,:] = np.array(sio.loadmat(data_dir + '10002'+str(i)+'/3/pca_mean.mat')['dist'])
    pca_m100[i-1,:] = np.array(sio.loadmat(data_dir + '10002'+str(i)+'/4/pca_mean.mat')['dist'])

    isomap_m10[i-1,:] = np.array(sio.loadmat(data_dir + '10002'+str(i)+'/1/isomap_mean.mat')['dist'])
    isomap_m40[i-1,:] = np.array(sio.loadmat(data_dir + '10002'+str(i)+'/2/isomap_mean.mat')['dist'])
    isomap_m70[i-1,:] = np.array(sio.loadmat(data_dir + '10002'+str(i)+'/3/isomap_mean.mat')['dist'])
    isomap_m100[i-1,:] = np.array(sio.loadmat(data_dir + '10002'+str(i)+'/4/isomap_mean.mat')['dist'])
```

```python
nl_10 = pca_m10/isomap_m10
nl_40 = pca_m40/isomap_m40
nl_70 = pca_m70/isomap_m70
nl_100 = pca_m100/isomap_m100
```

```python

x=range(5,60,6)
y = savgol_filter(np.mean(nl_10,axis=0),5,1)
error = stats.sem(nl_10,axis=0)
plt.plot(x, y, color='plum',linewidth=3, alpha=1.0)
plt.fill_between(x, y-error, y+error, color='plum', alpha=0.5)

x=range(5,60,6)
y = savgol_filter(np.mean(nl_40,axis=0),5,1)
error = stats.sem(nl_40,axis=0)
plt.plot(x, y, color='mediumpurple',linewidth=3, alpha=1.0)
plt.fill_between(x, y-error, y+error, color='mediumpurple', alpha=0.5)

x=range(5,60,6)
y = savgol_filter(np.mean(nl_70,axis=0),5,1)
error = stats.sem(nl_70,axis=0)
plt.plot(x, y, color='teal',linewidth=3, alpha=1.0)
plt.fill_between(x, y-error, y+error, color='teal', alpha=0.5)

x=range(5,60,6)
y = savgol_filter(np.mean(nl_100,axis=0),5,1)
error = stats.sem(nl_100,axis=0)
plt.plot(x, y, color='darkslategray',linewidth=3, alpha=1.0)
plt.fill_between(x, y-error, y+error, color='darkslategray', alpha=0.5)

plt.savefig('/home/cf620/Documents/curved_manifold/plots/standard_model_nl.pdf')
```
